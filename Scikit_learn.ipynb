## Fill missing data and transforming categorical data into numeric using scikit learn
 Step to do:
 1) Split your data first into(train/test)
 2) fill/transform the test and training data separately

```python
# import our missing car_sales data
car_sales_missing=pd.read_csv("car-sales-extended-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.isna().sum()
```

```python
# Drop the rows with no labels
car_sales_missing.dropna(subset=["Price"],inplace=True)
car_sales_missing.isna().sum()
```

```python
# Split X,y
X=car_sales_missing.drop("Price",axis=1)
y=car_sales_missing["Price"]

# from sklearn import train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X.isna().sum()
```

```python
# fill misisng value with scikit learn
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer

# fill categorical value with the missing value
cat_imputer=SimpleImputer(strategy="constant",fill_value="missing")
door_imputer=SimpleImputer(strategy="constant",fill_value=4)
num_imputer=SimpleImputer(strategy="mean")

# Define Columns
cat_features=["Make","Colour"]
door_features=["Doors"]
num_features=["Odometer (KM)"]

#Create an imputer(Something that fill missing values)
imputer=ColumnTransformer([("cat_imputer",cat_imputer,cat_features),
                           ("door_imputer",door_imputer,door_features),
                           ("num_imputer",num_imputer,num_features)])

# fill Train and test value separately

filled_X_train=imputer.fit_transform(X_train)
filled_X_test=imputer.fit_transform(X_test)

# Check filled X_train
filled_X_train
```

```python
# get our transformed data array back into DataFrames

car_sales_filled_train=pd.DataFrame(filled_X_train,columns=["Make","Colour","Doors","Odometer (KM)"])


car_sales_filled_test=pd.DataFrame(filled_X_test,columns=["Make","Colour","Doors","Odometer (KM)"])

# Check missing data in the train data

car_sales_filled_train.isna().sum()
                                    
```

```python
# lets one code all the data
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features=["Make","Colour","Doors"]
one_hot=OneHotEncoder()
transformer=ColumnTransformer([("one_hot",one_hot,categorical_features)],
                                remainder="passthrough")

transformed_X_train=transformer.fit_transform(car_sales_filled_train)
transformed_X_test=transformer.fit_transform(car_sales_filled_test)
```

```python
# Check transformed and fill transformed_X_train
transformed_X_train.toarray()
```

```python
# Now we have transformed X lets see we can fit our model or not
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor()
model.fit(transformed_X_train,y_train)
model.score(transformed_X_test,y_test)
```

## Picking machine learning model for regression problem
LET US USE California housing data set

```python
from sklearn.datasets import fetch_california_housing
housing=fetch_california_housing()
housing
```

```python
import pandas as pd
housing_df=pd.DataFrame(housing["data"],columns=housing["feature_names"])
housing_df
```

```python
housing_df["target"]=housing["target"]
housing_df.head()
```

```python
# lets check if any missing data is there or not
housing_df.isna().sum()
```

```python
# lets check wheather all tha data is in numeric form
housing_df.dtypes
```

```python
# Import algorith/estimator
from sklearn.linear_model import Ridge
# set up random seed
np.random.seed(42)

# Create X and y
X=housing_df.drop(["target"],axis=1)
y=housing_df["target"]

# import train_test_split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

# create a model
model=Ridge()

# fit the model
model.fit(X_train,y_train)

# check the score of model
model.score(X_train,y_train)
```

```python
# lets check the score on our test data
model.score(X_test,y_test)
```

# what if Ridge do not match our expectation
 Lets try some other algorithm like ensemble model which is combination of smaller model it is best to try and predict best output

```python
# importing another model for same dta
from sklearn.ensemble import RandomForestRegressor

# set the random_seed
np.random.seed(42)

# Create X and y
X=housing_df.drop(["target"],axis=1)
y=housing_df["target"]

#import train_test_Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)

# set the model
model=RandomForestRegressor()

# fit the model
model.fit(X_train,y_train)

# lets check the score on training data
model.score(X_train,y_train)
```

```python
# lets check on the test data
model.score(X_test,y_test)
```
